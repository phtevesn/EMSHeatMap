```python
import pandas as pd

#[latitude, logitude]
DOWNTOWN_COORD = [37.7705, -122.4269]
AIRPORT_COORD = [37.61962, -122.36562]
```

```python
downtown = pd.read_csv("sanfranciscodowntown.csv")
airport = pd.read_csv("sanfranciscointernationalairport.csv")
```

```python
downtown["latitude"] = DOWNTOWN_COORD[0]
downtown["longitude"] = DOWNTOWN_COORD[1]
airport["latitude"] = AIRPORT_COORD[0]
airport["longitude"] = AIRPORT_COORD[1]
```

```python
downtown.drop("TAVG (Degrees Fahrenheit)", axis=1, inplace=True)
airport.drop("TAVG (Degrees Fahrenheit)", axis=1, inplace=True)
```

```python
print(downtown.columns)
print(airport.columns)
```

```python
weather = pd.concat([downtown, airport], ignore_index=True)
```
